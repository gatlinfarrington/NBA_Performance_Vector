# Data Loading

```python
# Load the three datasets
player_stats = pd.read_csv('PlayerStatistics.csv')
players_info = pd.read_csv('Players.csv')
injury_data = pd.read_csv('InjuryStats.csv')

print(f"✓ Player Statistics: {player_stats.shape[0]:,} rows, {player_stats.shape[1]} columns")
print(f"✓ Player Info: {players_info.shape[0]:,} rows, {players_info.shape[1]} columns")
print(f"✓ Injury Data: {injury_data.shape[0]:,} rows, {injury_data.shape[1]} columns")
```

# Data Exploration

```python
# Convert dates to datetime and strip timezone info to avoid comparison issues
player_stats['gameDate'] = pd.to_datetime(player_stats['gameDate'], format='mixed', utc=True).dt.tz_localize(None)
injury_data['Date'] = pd.to_datetime(injury_data['Date'])

# Check date ranges
print(f"\nPlayer Statistics Date Range: {player_stats['gameDate'].min()} to {player_stats['gameDate'].max()}")
print(f"Injury Data Date Range: {injury_data['Date'].min()} to {injury_data['Date'].max()}")

# Filter to 2018-2023 (injury data constraint)
player_stats = player_stats[
    (player_stats['gameDate'] >= '2018-01-01') & 
    (player_stats['gameDate'] <= '2023-12-31')
]
print(f"\n✓ Filtered to 2018-2023: {player_stats.shape[0]:,} games")

# Check for missing values
print("\n--- Missing Values in Player Statistics ---")
print(player_stats.isnull().sum()[player_stats.isnull().sum() > 0])

# Display sample statistics
print("\n--- Sample Player Statistics ---")
print(player_stats[['firstName', 'lastName', 'gameDate', 'points', 'assists', 'reboundsTotal', 'numMinutes']].head())

print("\n--- Sample Injury Data ---")
print(injury_data[['Date', 'Team', 'Relinquished', 'Notes']].head(10))
```

# Feature Engineering for A_t

```python
print("\n[3] ENGINEERING FEATURES FOR AVAILABILITY (A_t)...")

# Create player full name for matching
player_stats['playerName'] = player_stats['firstName'] + ' ' + player_stats['lastName']
players_info['playerName'] = players_info['firstName'] + ' ' + players_info['lastName']

# CRITICAL: Handle missing values
print(f"  → Games with missing minutes: {player_stats['numMinutes'].isna().sum():,}")

# Work with only active games (where player actually played)
player_stats_active = player_stats[player_stats['numMinutes'].notna()].copy()
print(f"  → Active games (with minutes played): {len(player_stats_active):,}")

# Merge with player metadata (only if not already merged)
if 'height_y' not in player_stats_active.columns:
    player_stats_active = player_stats_active.merge(
        players_info[['personId', 'birthdate', 'height', 'bodyWeight', 'guard', 'forward', 'center']],
        on='personId',
        how='left',
        suffixes=('', '_meta')
    )

# Use the correct birthdate column (from players_info merge)
birthdate_col = 'birthdate_y' if 'birthdate_y' in player_stats_active.columns else 'birthdate'
height_col = 'height_y' if 'height_y' in player_stats_active.columns else 'height'
weight_col = 'bodyWeight_y' if 'bodyWeight_y' in player_stats_active.columns else 'bodyWeight'

# Calculate age at game time
player_stats_active[birthdate_col] = pd.to_datetime(player_stats_active[birthdate_col])
player_stats_active['age'] = (player_stats_active['gameDate'] - player_stats_active[birthdate_col]).dt.days / 365.25

# Standardize column names (remove _x and _y suffixes)
rename_dict = {}
if 'height_y' in player_stats_active.columns:
    rename_dict.update({
        'height_y': 'height',
        'bodyWeight_y': 'bodyWeight', 
        'guard_y': 'guard',
        'forward_y': 'forward',
        'center_y': 'center'
    })
    # Drop the _x versions if they exist
    drop_cols = ['height_x', 'bodyWeight_x', 'guard_x', 'forward_x', 'center_x', 'birthdate_x', 'birthdate_y']
    player_stats_active = player_stats_active.drop(columns=[c for c in drop_cols if c in player_stats_active.columns])
    player_stats_active = player_stats_active.rename(columns=rename_dict)

print(f"  ✓ Player metadata merged, age calculated")

# Sort by player and date for rolling calculations
player_stats_active = player_stats_active.sort_values(['personId', 'gameDate'])

# WORKLOAD FEATURES
print("  → Calculating workload metrics...")
player_stats_active['gameNumber'] = player_stats_active.groupby('personId').cumcount() + 1

# Rolling minutes (workload)
player_stats_active['minutes_last_5'] = player_stats_active.groupby('personId')['numMinutes'].transform(
    lambda x: x.rolling(5, min_periods=1).mean()
)
player_stats_active['minutes_last_10'] = player_stats_active.groupby('personId')['numMinutes'].transform(
    lambda x: x.rolling(10, min_periods=1).mean()
)

# Days since last game (rest)
player_stats_active['daysSinceLastGame'] = player_stats_active.groupby('personId')['gameDate'].diff().dt.days

# Back-to-back games indicator
player_stats_active['isBackToBack'] = (player_stats_active['daysSinceLastGame'] == 1).astype(int)

# Games in last 7/14 days - count games within time window
def count_games_in_window(group, window_days):
    """Count games in rolling time window"""
    result = []
    dates = group['gameDate'].values
    
    for i, current_date in enumerate(dates):
        cutoff_date = current_date - pd.Timedelta(days=window_days)
        games_in_window = np.sum((dates[:i] >= cutoff_date) & (dates[:i] < current_date))
        result.append(games_in_window)
    
    return result

print("  → Counting games in 7/14 day windows (this may take a moment)...")
player_stats_active['gamesInLast7Days'] = player_stats_active.groupby('personId').apply(
    lambda g: pd.Series(count_games_in_window(g, 7), index=g.index)
).reset_index(level=0, drop=True)

player_stats_active['gamesInLast14Days'] = player_stats_active.groupby('personId').apply(
    lambda g: pd.Series(count_games_in_window(g, 14), index=g.index)
).reset_index(level=0, drop=True)

print(f"  ✓ Workload features created")

# INJURY HISTORY FEATURES
print("  → Processing injury history...")

# Clean injury data
injury_data_clean = injury_data[injury_data['Relinquished'].notna()].copy()
injury_data_clean['playerName'] = injury_data_clean['Relinquished'].str.strip()

# Filter to our date range
injury_data_clean = injury_data_clean[
    (injury_data_clean['Date'] >= '2018-01-01') & 
    (injury_data_clean['Date'] <= '2023-12-31')
]

print(f"  → Found {len(injury_data_clean):,} injury events in 2018-2023")

# Initialize injury features
player_stats_active['priorInjuryCount'] = 0
player_stats_active['daysSinceLastInjury'] = np.nan

print("  → Calculating injury history for each player-game (this may take a few minutes)...")
unique_players = player_stats_active['playerName'].unique()

for idx, player_name in enumerate(unique_players):
    if idx % 100 == 0:
        print(f"     Processing player {idx+1}/{len(unique_players)}...")
    
    player_mask = player_stats_active['playerName'] == player_name
    player_injuries = injury_data_clean[injury_data_clean['playerName'] == player_name]['Date'].values
    
    if len(player_injuries) > 0:
        for row_idx in player_stats_active[player_mask].index:
            game_date = player_stats_active.loc[row_idx, 'gameDate']
            
            # Count injuries before this game
            prior_injuries = np.sum(player_injuries < game_date)
            player_stats_active.loc[row_idx, 'priorInjuryCount'] = prior_injuries
            
            # Days since most recent injury
            past_injuries = player_injuries[player_injuries < game_date]
            if len(past_injuries) > 0:
                days_since = (game_date - past_injuries[-1]).days
                player_stats_active.loc[row_idx, 'daysSinceLastInjury'] = days_since

print(f"  ✓ Injury history features created")
print(f"  → Games with injury history: {(player_stats_active['priorInjuryCount'] > 0).sum():,}")
```

# Feature Engineering for E_t

```python
print("\n[4] ENGINEERING FEATURES FOR EXECUTION (E_t)...")

# PERFORMANCE METRICS
print("  → Calculating performance metrics...")

# Fill NaN values in shooting stats with 0 (didn't attempt = didn't make)
player_stats_active['fieldGoalsMade'] = player_stats_active['fieldGoalsMade'].fillna(0)
player_stats_active['threePointersMade'] = player_stats_active['threePointersMade'].fillna(0)
player_stats_active['freeThrowsMade'] = player_stats_active['freeThrowsMade'].fillna(0)
player_stats_active['turnovers'] = player_stats_active['turnovers'].fillna(0)
player_stats_active['fieldGoalsAttempted'] = player_stats_active['fieldGoalsAttempted'].fillna(0)

# Player Efficiency Rating (simplified version)
# PER = (Points + Rebounds + Assists + Steals + Blocks - Turnovers - Missed FG) / Minutes
player_stats_active['PER_simple'] = (
    player_stats_active['points'] + 
    player_stats_active['reboundsTotal'] + 
    player_stats_active['assists'] + 
    player_stats_active['steals'] + 
    player_stats_active['blocks'] - 
    player_stats_active['turnovers'] - 
    (player_stats_active['fieldGoalsAttempted'] - player_stats_active['fieldGoalsMade'])
) / player_stats_active['numMinutes'].replace(0, np.nan)

# Fill inf values (division by zero) with NaN
player_stats_active['PER_simple'] = player_stats_active['PER_simple'].replace([np.inf, -np.inf], np.nan)

print(f"  → Created PER metric (non-null: {player_stats_active['PER_simple'].notna().sum():,})")

# Rolling performance averages
print("  → Calculating rolling performance averages...")
for stat in ['points', 'assists', 'reboundsTotal', 'PER_simple', 'plusMinusPoints']:
    # Last 5 games
    player_stats_active[f'{stat}_last_5'] = player_stats_active.groupby('personId')[stat].transform(
        lambda x: x.rolling(5, min_periods=1).mean()
    )
    # Last 10 games
    player_stats_active[f'{stat}_last_10'] = player_stats_active.groupby('personId')[stat].transform(
        lambda x: x.rolling(10, min_periods=1).mean()
    )

# Performance volatility (standard deviation of last 10 games)
print("  → Calculating performance volatility...")
player_stats_active['points_volatility'] = player_stats_active.groupby('personId')['points'].transform(
    lambda x: x.rolling(10, min_periods=3).std()
)

print(f"  ✓ Performance features created")

# CONTEXTUAL FEATURES
print("  → Adding contextual features...")

# Home/away
player_stats_active['isHome'] = player_stats_active['home'].astype(int)

# Season extraction
player_stats_active['season'] = player_stats_active['gameDate'].dt.year

# Position dummies (already have guard, forward, center from merge)

print(f"  ✓ Contextual features added")

# Check for any remaining issues
print(f"\n  Data quality check:")
print(f"     Total games: {len(player_stats_active):,}")
print(f"     Games with PER: {player_stats_active['PER_simple'].notna().sum():,}")
print(f"     Games with points_last_5: {player_stats_active['points_last_5'].notna().sum():,}")
print(f"     Games with volatility: {player_stats_active['points_volatility'].notna().sum():,}")
```

# Create Modeling Datasets

```python
print("\n[5] PREPARING MODELING DATASETS...")

# Remove rows with too many NaN values (early games with insufficient history)
player_stats_clean = player_stats_active.dropna(subset=['minutes_last_5', 'points_last_5', 'daysSinceLastGame'])

print(f"  → Clean dataset: {player_stats_clean.shape[0]:,} games")

# AVAILABILITY (A_t) DATASET
print("\n  → Creating Availability (A_t) target...")
# Target: Will player be injured in next N days?
INJURY_PREDICTION_WINDOW = 14  # Predict injuries in next 14 days

# Create forward-looking injury label
player_stats_clean = player_stats_clean.sort_values(['personId', 'gameDate'])
player_stats_clean['futureInjury'] = 0

print(f"  → Calculating future injury labels (14-day window)...")
unique_players = player_stats_clean['playerName'].unique()

for idx, player_name in enumerate(unique_players):
    if idx % 100 == 0:
        print(f"     Processing player {idx+1}/{len(unique_players)}...")
    
    player_mask = player_stats_clean['playerName'] == player_name
    player_injuries = injury_data_clean[injury_data_clean['playerName'] == player_name]['Date'].values
    
    if len(player_injuries) > 0:
        for row_idx in player_stats_clean[player_mask].index:
            game_date = player_stats_clean.loc[row_idx, 'gameDate']
            future_date = game_date + timedelta(days=INJURY_PREDICTION_WINDOW)
            
            # Check if there's an injury between game_date and future_date
            future_injuries = player_injuries[(player_injuries > game_date) & (player_injuries <= future_date)]
            if len(future_injuries) > 0:
                player_stats_clean.loc[row_idx, 'futureInjury'] = 1

print(f"  ✓ Future injury labels created")

# Define features for Availability model
availability_features = [
    'age', 'bodyWeight', 'height',
    'minutes_last_5', 'minutes_last_10',
    'daysSinceLastGame', 'isBackToBack',
    'gamesInLast7Days', 'gamesInLast14Days',
    'priorInjuryCount', 'daysSinceLastInjury',
    'points_last_5', 'PER_simple'  # Performance can indicate fatigue
]

# EXECUTION (E_t) DATASET
print("\n  → Creating Execution (E_t) target...")
# Target: Predict next game performance (points)
player_stats_clean['nextGamePoints'] = player_stats_clean.groupby('personId')['points'].shift(-1)
player_stats_clean['nextGamePER'] = player_stats_clean.groupby('personId')['PER_simple'].shift(-1)

print(f"  ✓ Next game performance targets created")

# Define features for Execution model
execution_features = [
    'age', 'guard', 'forward', 'center',
    'points_last_5', 'points_last_10', 'points_volatility',
    'assists_last_5', 'reboundsTotal_last_5',
    'PER_simple_last_5', 'PER_simple_last_10',
    'plusMinusPoints_last_5',
    'daysSinceLastGame', 'isBackToBack',
    'minutes_last_5', 'isHome'
]

# Remove rows where targets or critical features are missing
print("\n  → Filtering datasets...")
availability_data = player_stats_clean.dropna(subset=availability_features + ['futureInjury']).copy()
execution_data = player_stats_clean.dropna(subset=execution_features + ['nextGamePoints']).copy()

print(f"\n✓ Availability dataset: {availability_data.shape[0]:,} samples")
print(f"  - Injury rate: {availability_data['futureInjury'].mean():.2%}")
print(f"  - Players with injuries in next 14 days: {availability_data['futureInjury'].sum():,}")

print(f"\n✓ Execution dataset: {execution_data.shape[0]:,} samples")
print(f"  - Avg next game points: {execution_data['nextGamePoints'].mean():.1f}")
print(f"  - Std next game points: {execution_data['nextGamePoints'].std():.1f}")

# Quick sanity check
print(f"\n  → Sanity checks:")
print(f"     Availability features all present: {all(f in availability_data.columns for f in availability_features)}")
print(f"     Execution features all present: {all(f in execution_data.columns for f in execution_features)}")
```

# train/test Split

```python
print("\n[6] CREATING TEMPORAL TRAIN/TEST SPLITS...")

# Use 2018-2022 for training, 2023 for testing (respects temporal ordering)
SPLIT_DATE = '2023-01-01'

print(f"  → Split date: {SPLIT_DATE}")
print(f"  → Training: 2018-2022")
print(f"  → Testing: 2023")

# AVAILABILITY SPLIT
availability_train = availability_data[availability_data['gameDate'] < SPLIT_DATE]
availability_test = availability_data[availability_data['gameDate'] >= SPLIT_DATE]

X_train_avail = availability_train[availability_features]
y_train_avail = availability_train['futureInjury']
X_test_avail = availability_test[availability_features]
y_test_avail = availability_test['futureInjury']

print(f"\nAvailability (A_t):")
print(f"  Train: {len(X_train_avail):,} samples (injury rate: {y_train_avail.mean():.2%})")
print(f"  Test:  {len(X_test_avail):,} samples (injury rate: {y_test_avail.mean():.2%})")

# Check for class imbalance
print(f"  Class distribution in training:")
print(f"    No injury (0): {(y_train_avail == 0).sum():,} ({(y_train_avail == 0).mean():.1%})")
print(f"    Injury (1):    {(y_train_avail == 1).sum():,} ({(y_train_avail == 1).mean():.1%})")

# EXECUTION SPLIT
execution_train = execution_data[execution_data['gameDate'] < SPLIT_DATE]
execution_test = execution_data[execution_data['gameDate'] >= SPLIT_DATE]

X_train_exec = execution_train[execution_features]
y_train_exec = execution_train['nextGamePoints']
X_test_exec = execution_test[execution_features]
y_test_exec = execution_test['nextGamePoints']

print(f"\nExecution (E_t):")
print(f"  Train: {len(X_train_exec):,} samples")
print(f"  Test:  {len(X_test_exec):,} samples")
print(f"  Train avg points: {y_train_exec.mean():.1f} ± {y_train_exec.std():.1f}")
print(f"  Test avg points:  {y_test_exec.mean():.1f} ± {y_test_exec.std():.1f}")

# Store metadata for later analysis
availability_train_meta = availability_train[['personId', 'playerName', 'gameDate', 'futureInjury']]
availability_test_meta = availability_test[['personId', 'playerName', 'gameDate', 'futureInjury']]
execution_train_meta = execution_train[['personId', 'playerName', 'gameDate', 'nextGamePoints']]
execution_test_meta = execution_test[['personId', 'playerName', 'gameDate', 'nextGamePoints']]

print("\n✓ Train/test splits created")
```

# Availability Models

```python
print("\n" + "=" * 80)
print("BUILDING AVAILABILITY (A_t) MODELS")
print("=" * 80)

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.preprocessing import StandardScaler

# Scale features for models that need it
print("\n[8.1] Preprocessing features...")
scaler_avail = StandardScaler()
X_train_avail_scaled = scaler_avail.fit_transform(X_train_avail)
X_test_avail_scaled = scaler_avail.transform(X_test_avail)

print("\n[8.2] Training Logistic Regression (Baseline)...")
lr_model = LogisticRegression(
    random_state=42,
    max_iter=1000,
    class_weight='balanced'  # Handle class imbalance
)
lr_model.fit(X_train_avail_scaled, y_train_avail)

# Predictions
lr_pred_train = lr_model.predict(X_train_avail_scaled)
lr_pred_test = lr_model.predict(X_test_avail_scaled)
lr_pred_proba_test = lr_model.predict_proba(X_test_avail_scaled)[:, 1]

# Evaluation
lr_train_auc = roc_auc_score(y_train_avail, lr_model.predict_proba(X_train_avail_scaled)[:, 1])
lr_test_auc = roc_auc_score(y_test_avail, lr_pred_proba_test)

print(f"  Train AUC: {lr_train_auc:.4f}")
print(f"  Test AUC:  {lr_test_auc:.4f}")
print(f"\n  Classification Report (Test):")
print(classification_report(y_test_avail, lr_pred_test, target_names=['No Injury', 'Injury']))

print("\n[8.3] Training Random Forest...")
rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    min_samples_split=50,
    min_samples_leaf=20,
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)
rf_model.fit(X_train_avail, y_train_avail)

# Predictions
rf_pred_train = rf_model.predict(X_train_avail)
rf_pred_test = rf_model.predict(X_test_avail)
rf_pred_proba_test = rf_model.predict_proba(X_test_avail)[:, 1]

# Evaluation
rf_train_auc = roc_auc_score(y_train_avail, rf_model.predict_proba(X_train_avail)[:, 1])
rf_test_auc = roc_auc_score(y_test_avail, rf_pred_proba_test)

print(f"  Train AUC: {rf_train_auc:.4f}")
print(f"  Test AUC:  {rf_test_auc:.4f}")
print(f"\n  Classification Report (Test):")
print(classification_report(y_test_avail, rf_pred_test, target_names=['No Injury', 'Injury']))

print("\n[8.4] Training XGBoost...")

# Calculate scale_pos_weight for class imbalance
scale_pos_weight = (y_train_avail == 0).sum() / (y_train_avail == 1).sum()

xgb_model = XGBClassifier(
    n_estimators=100,
    max_depth=6,
    learning_rate=0.1,
    scale_pos_weight=scale_pos_weight,
    random_state=42,
    n_jobs=-1,
    eval_metric='logloss'
)
xgb_model.fit(X_train_avail, y_train_avail)

# Predictions
xgb_pred_train = xgb_model.predict(X_train_avail)
xgb_pred_test = xgb_model.predict(X_test_avail)
xgb_pred_proba_test = xgb_model.predict_proba(X_test_avail)[:, 1]

# Evaluation
xgb_train_auc = roc_auc_score(y_train_avail, xgb_model.predict_proba(X_train_avail)[:, 1])
xgb_test_auc = roc_auc_score(y_test_avail, xgb_pred_proba_test)

print(f"  Train AUC: {xgb_train_auc:.4f}")
print(f"  Test AUC:  {xgb_test_auc:.4f}")
print(f"\n  Classification Report (Test):")
print(classification_report(y_test_avail, xgb_pred_test, target_names=['No Injury', 'Injury']))

print("\n" + "=" * 80)
print("AVAILABILITY MODEL COMPARISON")
print("=" * 80)

comparison_df = pd.DataFrame({
    'Model': ['Logistic Regression', 'Random Forest', 'XGBoost'],
    'Train AUC': [lr_train_auc, rf_train_auc, xgb_train_auc],
    'Test AUC': [lr_test_auc, rf_test_auc, xgb_test_auc],
    'Overfit Gap': [lr_train_auc - lr_test_auc, rf_train_auc - rf_test_auc, xgb_train_auc - xgb_test_auc]
})
print(comparison_df.to_string(index=False))

best_model_name = comparison_df.loc[comparison_df['Test AUC'].idxmax(), 'Model']
print(f"\n✓ Best model: {best_model_name}")

# Store best model predictions
if best_model_name == 'XGBoost':
    best_avail_pred = xgb_pred_proba_test
    best_avail_model = xgb_model
elif best_model_name == 'Random Forest':
    best_avail_pred = rf_pred_proba_test
    best_avail_model = rf_model
else:
    best_avail_pred = lr_pred_proba_test
    best_avail_model = lr_model

print("=" * 80)
```

# Execution Models

```python
print("\n" + "=" * 80)
print("BUILDING EXECUTION (E_t) MODELS")
print("=" * 80)

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Scale features for Linear Regression
print("\n[9.1] Preprocessing features...")
scaler_exec = StandardScaler()
X_train_exec_scaled = scaler_exec.fit_transform(X_train_exec)
X_test_exec_scaled = scaler_exec.transform(X_test_exec)

print(f"  ✓ Features scaled")

print("\n[9.2] Training Linear Regression (Baseline)...")
lr_exec_model = LinearRegression()
lr_exec_model.fit(X_train_exec_scaled, y_train_exec)

# Predictions
lr_exec_pred_train = lr_exec_model.predict(X_train_exec_scaled)
lr_exec_pred_test = lr_exec_model.predict(X_test_exec_scaled)

# Evaluation
lr_exec_train_r2 = r2_score(y_train_exec, lr_exec_pred_train)
lr_exec_test_r2 = r2_score(y_test_exec, lr_exec_pred_test)
lr_exec_train_rmse = np.sqrt(mean_squared_error(y_train_exec, lr_exec_pred_train))
lr_exec_test_rmse = np.sqrt(mean_squared_error(y_test_exec, lr_exec_pred_test))
lr_exec_test_mae = mean_absolute_error(y_test_exec, lr_exec_pred_test)

print(f"  Train R²:   {lr_exec_train_r2:.4f}")
print(f"  Test R²:    {lr_exec_test_r2:.4f}")
print(f"  Train RMSE: {lr_exec_train_rmse:.2f}")
print(f"  Test RMSE:  {lr_exec_test_rmse:.2f}")
print(f"  Test MAE:   {lr_exec_test_mae:.2f}")

print("\n[9.3] Training Random Forest Regressor...")
rf_exec_model = RandomForestRegressor(
    n_estimators=100,
    max_depth=15,
    min_samples_split=20,
    min_samples_leaf=10,
    random_state=42,
    n_jobs=-1
)
rf_exec_model.fit(X_train_exec, y_train_exec)

# Predictions
rf_exec_pred_train = rf_exec_model.predict(X_train_exec)
rf_exec_pred_test = rf_exec_model.predict(X_test_exec)

# Evaluation
rf_exec_train_r2 = r2_score(y_train_exec, rf_exec_pred_train)
rf_exec_test_r2 = r2_score(y_test_exec, rf_exec_pred_test)
rf_exec_train_rmse = np.sqrt(mean_squared_error(y_train_exec, rf_exec_pred_train))
rf_exec_test_rmse = np.sqrt(mean_squared_error(y_test_exec, rf_exec_pred_test))
rf_exec_test_mae = mean_absolute_error(y_test_exec, rf_exec_pred_test)

print(f"  Train R²:   {rf_exec_train_r2:.4f}")
print(f"  Test R²:    {rf_exec_test_r2:.4f}")
print(f"  Train RMSE: {rf_exec_train_rmse:.2f}")
print(f"  Test RMSE:  {rf_exec_test_rmse:.2f}")
print(f"  Test MAE:   {rf_exec_test_mae:.2f}")

print("\n" + "=" * 80)
print("EXECUTION MODEL COMPARISON")
print("=" * 80)

exec_comparison_df = pd.DataFrame({
    'Model': ['Linear Regression', 'Random Forest'],
    'Train R²': [lr_exec_train_r2, rf_exec_train_r2],
    'Test R²': [lr_exec_test_r2, rf_exec_test_r2],
    'Test RMSE': [lr_exec_test_rmse, rf_exec_test_rmse],
    'Test MAE': [lr_exec_test_mae, rf_exec_test_mae],
    'Overfit (R² gap)': [lr_exec_train_r2 - lr_exec_test_r2, rf_exec_train_r2 - rf_exec_test_r2]
})
print(exec_comparison_df.to_string(index=False))

best_exec_model_name = exec_comparison_df.loc[exec_comparison_df['Test R²'].idxmax(), 'Model']
print(f"\n✓ Best model: {best_exec_model_name}")

# Store best model predictions
if best_exec_model_name == 'Random Forest':
    best_exec_pred = rf_exec_pred_test
    best_exec_model = rf_exec_model
else:
    best_exec_pred = lr_exec_pred_test
    best_exec_model = lr_exec_model

# Quick interpretation check
print(f"\n📊 Prediction Quality:")
print(f"   Baseline (always predict mean): RMSE = {y_test_exec.std():.2f}")
print(f"   Best model RMSE: {exec_comparison_df.loc[exec_comparison_df['Test R²'].idxmax(), 'Test RMSE']:.2f}")
improvement = (1 - exec_comparison_df.loc[exec_comparison_df['Test R²'].idxmax(), 'Test RMSE'] / y_test_exec.std()) * 100
print(f"   Improvement over baseline: {improvement:.1f}%")

print("=" * 80)
```

# Combining A_t and E_t to form P_t

```python
print("\n" + "=" * 80)
print("CREATING PERFORMANCE VECTOR P_t = [A_t, E_t]")
print("=" * 80)

print("\n[10.1] Aligning Availability and Execution predictions...")

# Get common games in test set
avail_test_ids = set(zip(availability_test_meta['personId'], availability_test_meta['gameDate']))
exec_test_ids = set(zip(execution_test_meta['personId'], execution_test_meta['gameDate']))
common_test_ids = avail_test_ids.intersection(exec_test_ids)

print(f"  → Availability test samples: {len(avail_test_ids):,}")
print(f"  → Execution test samples: {len(exec_test_ids):,}")
print(f"  → Common samples (both A_t and E_t): {len(common_test_ids):,}")

# Filter to common samples
avail_test_mask = availability_test_meta.apply(
    lambda row: (row['personId'], row['gameDate']) in common_test_ids, axis=1
)
exec_test_mask = execution_test_meta.apply(
    lambda row: (row['personId'], row['gameDate']) in common_test_ids, axis=1
)

# Create aligned datasets
aligned_avail_meta = availability_test_meta[avail_test_mask].copy()
aligned_exec_meta = execution_test_meta[exec_test_mask].copy()

# Add predictions to metadata for easier sorting
aligned_avail_meta['pred'] = best_avail_pred[avail_test_mask.values]
aligned_avail_meta['actual'] = y_test_avail[avail_test_mask.values].values
aligned_exec_meta['pred'] = best_exec_pred[exec_test_mask.values]
aligned_exec_meta['actual'] = y_test_exec[exec_test_mask.values].values

# Sort both by personId and gameDate
aligned_avail_meta = aligned_avail_meta.sort_values(['personId', 'gameDate']).reset_index(drop=True)
aligned_exec_meta = aligned_exec_meta.sort_values(['personId', 'gameDate']).reset_index(drop=True)

# Verify alignment
assert (aligned_avail_meta['personId'] == aligned_exec_meta['personId']).all()
assert (aligned_avail_meta['gameDate'] == aligned_exec_meta['gameDate']).all()
print(f"\n  ✓ Datasets aligned: {len(aligned_avail_meta):,} samples")

# Extract predictions and actuals
aligned_avail_pred = aligned_avail_meta['pred'].values
aligned_exec_pred = aligned_exec_meta['pred'].values
aligned_y_avail = aligned_avail_meta['actual'].values
aligned_y_exec = aligned_exec_meta['actual'].values

print("\n[10.2] Normalizing predictions to [0, 1] scale...")

# A_t is already a probability (0-1) from predict_proba
A_t = aligned_avail_pred

# E_t: Normalize predicted points to [0, 1]
# Use min-max normalization based on realistic NBA scoring range (0-50 points)
E_t_raw = aligned_exec_pred
E_t = np.clip(E_t_raw / 50.0, 0, 1)  # Clip to [0, 1]

print(f"  A_t (availability prob): min={A_t.min():.3f}, max={A_t.max():.3f}, mean={A_t.mean():.3f}")
print(f"  E_t (execution, normalized): min={E_t.min():.3f}, max={E_t.max():.3f}, mean={E_t.mean():.3f}")

print("\n[10.3] Creating Performance Vector P_t = [A_t, E_t]...")

# Simple approach: weighted combination
# Weight availability more (injury risk is critical)
w_avail = 0.6
w_exec = 0.4

P_t = w_avail * A_t + w_exec * E_t

print(f"  Weights: A_t={w_avail}, E_t={w_exec}")
print(f"  P_t: min={P_t.min():.3f}, max={P_t.max():.3f}, mean={P_t.mean():.3f}")

performance_df = pd.DataFrame({
    'personId': aligned_avail_meta['personId'],
    'playerName': aligned_avail_meta['playerName'],
    'gameDate': aligned_avail_meta['gameDate'],
    'A_t_pred': A_t,
    'A_t_actual': 1 - aligned_y_avail,  # Convert injury flag to availability
    'E_t_pred': E_t_raw,  # Keep raw points for interpretability
    'E_t_actual': aligned_y_exec,
    'P_t': P_t
})

print(f"\n  ✓ Performance vector created for {len(performance_df):,} player-games")


print("\n[10.4] Performance Vector Analysis...")

# Top 10 by P_t
top_performers = performance_df.nlargest(10, 'P_t')
print("\n🏆 Top 10 Predicted Performers (by P_t):")
print(top_performers[['playerName', 'gameDate', 'A_t_pred', 'E_t_pred', 'P_t']].to_string(index=False))

# Bottom 10 by P_t
bottom_performers = performance_df.nsmallest(10, 'P_t')
print("\n⚠️  Bottom 10 Predicted Performers (by P_t):")
print(bottom_performers[['playerName', 'gameDate', 'A_t_pred', 'E_t_pred', 'P_t']].to_string(index=False))

# Aggregate by player (average P_t across all games)
player_avg_performance = performance_df.groupby('playerName').agg({
    'P_t': 'mean',
    'A_t_pred': 'mean',
    'E_t_pred': 'mean',
    'gameDate': 'count'
}).rename(columns={'gameDate': 'num_games'})

player_avg_performance = player_avg_performance[player_avg_performance['num_games'] >= 10]  # Min 10 games
top_players = player_avg_performance.nlargest(10, 'P_t')

print("\n🌟 Top 10 Players (avg P_t, min 10 games):")
print(top_players.to_string())

print("\n" + "=" * 80)
print("PERFORMANCE VECTOR CREATION COMPLETE")
print("=" * 80)
```

```python
print("\n" + "=" * 80)
print("ADVANCED ENSEMBLE: STACKED META-LEARNERS")
print("=" * 80)

from sklearn.ensemble import StackingClassifier, StackingRegressor
from sklearn.linear_model import LogisticRegression, Ridge

print("\n[12.1] Building Stacked Availability Model...")

# Define base models (use the ones we already trained)
base_models_avail = [
    ('logistic', LogisticRegression(random_state=42, max_iter=1000, class_weight='balanced')),
    ('random_forest', RandomForestClassifier(
        n_estimators=100, max_depth=10, min_samples_split=50, 
        min_samples_leaf=20, class_weight='balanced', random_state=42
    )),
    ('xgboost', XGBClassifier(
        n_estimators=100, max_depth=6, learning_rate=0.1,
        scale_pos_weight=(y_train_avail == 0).sum() / (y_train_avail == 1).sum(),
        random_state=42, eval_metric='logloss'
    ))
]

# Meta-learner: Logistic Regression (learns optimal weights for base models)
# FIXED: Remove n_jobs to avoid serialization issues
stacked_avail = StackingClassifier(
    estimators=base_models_avail,
    final_estimator=LogisticRegression(random_state=42),
    cv=5  # 5-fold cross-validation for base predictions
)

print("  → Training stacked model with 5-fold CV (this may take a few minutes)...")
stacked_avail.fit(X_train_avail_scaled, y_train_avail)

# Predictions
stacked_avail_pred_train = stacked_avail.predict_proba(X_train_avail_scaled)[:, 1]
stacked_avail_pred_test = stacked_avail.predict_proba(X_test_avail_scaled)[:, 1]
stacked_avail_pred_test_binary = stacked_avail.predict(X_test_avail_scaled)

# Evaluation
stacked_avail_train_auc = roc_auc_score(y_train_avail, stacked_avail_pred_train)
stacked_avail_test_auc = roc_auc_score(y_test_avail, stacked_avail_pred_test)

print(f"\n  Stacked Model Performance:")
print(f"    Train AUC: {stacked_avail_train_auc:.4f}")
print(f"    Test AUC:  {stacked_avail_test_auc:.4f}")
print(f"    Overfit gap: {stacked_avail_train_auc - stacked_avail_test_auc:.4f}")

print(f"\n  Classification Report (Test):")
print(classification_report(y_test_avail, stacked_avail_pred_test_binary, 
                           target_names=['No Injury', 'Injury']))

# Compare to best individual model
print(f"\n  Comparison to Best Individual Model (Logistic Regression):")
print(f"    Individual Test AUC: {lr_test_auc:.4f}")
print(f"    Stacked Test AUC:    {stacked_avail_test_auc:.4f}")
improvement = stacked_avail_test_auc - lr_test_auc
pct_improvement = (improvement / lr_test_auc * 100) if lr_test_auc > 0 else 0
print(f"    Improvement:         {improvement:.4f} ({pct_improvement:.2f}%)")

print("\n[12.2] Building Stacked Execution Model...")

# Define base models
base_models_exec = [
    ('linear', LinearRegression()),
    ('random_forest', RandomForestRegressor(
        n_estimators=100, max_depth=15, min_samples_split=20,
        min_samples_leaf=10, random_state=42
    ))
]

# Meta-learner: Ridge Regression (regularized to prevent overfitting)
stacked_exec = StackingRegressor(
    estimators=base_models_exec,
    final_estimator=Ridge(alpha=1.0),
    cv=5
)

print("  → Training stacked model with 5-fold CV...")
stacked_exec.fit(X_train_exec_scaled, y_train_exec)

# Predictions
stacked_exec_pred_train = stacked_exec.predict(X_train_exec_scaled)
stacked_exec_pred_test = stacked_exec.predict(X_test_exec_scaled)

# Evaluation
stacked_exec_train_r2 = r2_score(y_train_exec, stacked_exec_pred_train)
stacked_exec_test_r2 = r2_score(y_test_exec, stacked_exec_pred_test)
stacked_exec_train_rmse = np.sqrt(mean_squared_error(y_train_exec, stacked_exec_pred_train))
stacked_exec_test_rmse = np.sqrt(mean_squared_error(y_test_exec, stacked_exec_pred_test))
stacked_exec_test_mae = mean_absolute_error(y_test_exec, stacked_exec_pred_test)

print(f"\n  Stacked Model Performance:")
print(f"    Train R²:   {stacked_exec_train_r2:.4f}")
print(f"    Test R²:    {stacked_exec_test_r2:.4f}")
print(f"    Train RMSE: {stacked_exec_train_rmse:.2f}")
print(f"    Test RMSE:  {stacked_exec_test_rmse:.2f}")
print(f"    Test MAE:   {stacked_exec_test_mae:.2f}")

# Compare to best individual model
print(f"\n  Comparison to Best Individual Model (Linear Regression):")
print(f"    Individual Test R²:   {lr_exec_test_r2:.4f}")
print(f"    Stacked Test R²:      {stacked_exec_test_r2:.4f}")
print(f"    Improvement:          {stacked_exec_test_r2 - lr_exec_test_r2:.4f}")
print(f"    Individual Test RMSE: {lr_exec_test_rmse:.2f}")
print(f"    Stacked Test RMSE:    {stacked_exec_test_rmse:.2f}")
print(f"    RMSE Reduction:       {lr_exec_test_rmse - stacked_exec_test_rmse:.2f} points")

print("\n[12.3] Creating Updated Performance Vector with Stacked Predictions...")

# Get stacked predictions for the aligned test set
# Need to re-align since we're using different predictions
stacked_avail_test_pred_full = stacked_avail.predict_proba(X_test_avail_scaled)[:, 1]
stacked_exec_test_pred_full = stacked_exec.predict(X_test_exec_scaled)

# Use same alignment as before
aligned_stacked_avail_pred = stacked_avail_test_pred_full[avail_test_mask.values]
aligned_stacked_exec_pred = stacked_exec_test_pred_full[exec_test_mask.values]

# Sort to match alignment
aligned_stacked_avail_meta = availability_test_meta[avail_test_mask].copy()
aligned_stacked_exec_meta = execution_test_meta[exec_test_mask].copy()

aligned_stacked_avail_meta['pred'] = aligned_stacked_avail_pred
aligned_stacked_exec_meta['pred'] = aligned_stacked_exec_pred

aligned_stacked_avail_meta = aligned_stacked_avail_meta.sort_values(['personId', 'gameDate']).reset_index(drop=True)
aligned_stacked_exec_meta = aligned_stacked_exec_meta.sort_values(['personId', 'gameDate']).reset_index(drop=True)

# Verify alignment
assert (aligned_stacked_avail_meta['personId'] == aligned_stacked_exec_meta['personId']).all()
assert (aligned_stacked_avail_meta['gameDate'] == aligned_stacked_exec_meta['gameDate']).all()

# Extract aligned predictions
A_t_stacked = aligned_stacked_avail_meta['pred'].values
E_t_stacked_raw = aligned_stacked_exec_meta['pred'].values
E_t_stacked = np.clip(E_t_stacked_raw / 50.0, 0, 1)

P_t_stacked = w_avail * A_t_stacked + w_exec * E_t_stacked

print(f"  Stacked P_t: min={P_t_stacked.min():.3f}, max={P_t_stacked.max():.3f}, mean={P_t_stacked.mean():.3f}")

# Compare to original
print(f"\n  Comparison to Original P_t:")
print(f"    Original P_t mean: {P_t.mean():.3f}")
print(f"    Stacked P_t mean:  {P_t_stacked.mean():.3f}")
print(f"    Correlation:       {np.corrcoef(P_t, P_t_stacked)[0, 1]:.3f}")

# Update performance dataframe with stacked predictions
performance_df['A_t_pred_stacked'] = A_t_stacked
performance_df['E_t_pred_stacked'] = E_t_stacked_raw
performance_df['P_t_stacked'] = P_t_stacked

# Top performers with stacked model
top_performers_stacked = performance_df.nlargest(10, 'P_t_stacked')
print("\n🏆 Top 10 Performers (Stacked P_t):")
print(top_performers_stacked[['playerName', 'gameDate', 'A_t_pred_stacked', 'E_t_pred_stacked', 'P_t_stacked']].to_string(index=False))

# Player-level aggregation
player_avg_stacked = performance_df.groupby('playerName').agg({
    'P_t_stacked': 'mean',
    'A_t_pred_stacked': 'mean',
    'E_t_pred_stacked': 'mean',
    'gameDate': 'count'
}).rename(columns={'gameDate': 'num_games'})

player_avg_stacked = player_avg_stacked[player_avg_stacked['num_games'] >= 10]
top_players_stacked = player_avg_stacked.nlargest(10, 'P_t_stacked')

print("\n🌟 Top 10 Players (avg Stacked P_t, min 10 games):")
print(top_players_stacked.to_string())

print("\n" + "=" * 80)
print("STACKING COMPLETE")
print("=" * 80)
```
